```python
#Array Attributes
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

```python
#Array Operations
```

```python
# Basic Math
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
np.sin(a)
```

```python
# Matrix Multiplication
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
np.dot(A, B)
A @ B
```

```python
# Aggregation
```

```python
arr = np.array([1, 2, 3, 4])
print(arr.sum())
print(arr.mean())
print(arr.max())
print(arr.min())
print(arr.std())
```

```python
#Indexing and slicing
```

```python
# 1D Arrays
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr[0])
print(arr[1:4])
print(arr[::2])
```

```python
# 2D Arrays
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr[0, 1])
print(arr[:,1])
print(arr[1,:2])
```

```python
# Boolean Indexing
```

```python
arr=np.array([1,2,3,4,5])
mask=arr>3
arr[mask]
```
